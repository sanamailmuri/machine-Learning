 **What is a Support Vector Machine (SVM)**

SVM is a supervised machine learning algorithm used for classification and regression tasks. It finds the optimal hyperplane that best separates data points of different classes by maximizing the margin between the classes' closest points (support vectors).

**What is the difference between Hard Margin and Soft Margin SVM**

Hard Margin SVM assumes data is linearly separable and finds a hyperplane that perfectly separates classes with no misclassification.

Soft Margin SVM allows some misclassifications (violations) using slack variables to handle non-linearly separable or noisy data, balancing margin maximization and classification errors.

** What is the role of Lagrange Multipliers in SVM**

Lagrange multipliers transform the constrained optimization problem into a dual form, allowing optimization using only inner products of data points. This dual form makes kernel methods feasible for nonlinear decision boundaries.

** What are Support Vectors in SVM**

Support vectors are the training data points closest to the decision boundary (hyperplane) that define the margin. They are critical in determining the position and orientation of the hyperplane.

**What is a Support Vector Classifier (SVC)**

SVC is the implementation of SVM for classification tasks. It finds the hyperplane that best separates classes, potentially using kernels for nonlinear separations.

**What is a Support Vector Regressor (SVR)**

SVR applies SVM principles to regression problems, trying to fit a function within a margin of tolerance (epsilon) and minimizing deviations outside this margin.

** What is the Kernel Trick in SVM**

The kernel trick implicitly maps data into a higher-dimensional space without explicitly computing coordinates in that space, allowing linear separation of non-linearly separable data using kernel functions like polynomial, RBF, etc.

** Compare Linear Kernel, Polynomial Kernel, and RBF Kernel**

Linear Kernel: No mapping; good for linearly separable data.

Polynomial Kernel: Maps data into polynomial feature space; controls interaction features.

RBF Kernel: Maps data into infinite-dimensional space; handles complex boundaries well, sensitive to parameter gamma.

** What is the effect of the C parameter in SVM**

C controls the trade-off between maximizing the margin and minimizing classification errors. A small C allows a wider margin with more misclassifications (soft margin); a large C tries to classify all points correctly, possibly overfitting.

** What is the role of the Gamma parameter in RBF Kernel SVM**

Gamma defines the influence range of a single training example. Low gamma means far reach (smooth decision boundary), high gamma means close reach (complex boundary, potential overfitting).

** What is the Naïve Bayes classifier, and why is it called "Naïve**

Naïve Bayes is a probabilistic classifier based on Bayes' theorem with the "naïve" assumption that features are conditionally independent given the class label.

** Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes**

Gaussian NB: Assumes continuous features follow a normal distribution.

Multinomial NB: Used for discrete counts/features (e.g., word counts in text).

Bernoulli NB: Used for binary/boolean features (e.g., word presence or absence).

** When should you use Gaussian Naïve Bayes over other variants**

Use Gaussian NB for continuous numerical data where features approximately follow a normal distribution.

** What are the key assumptions made by Naïve Bayes**

Feature independence: each feature contributes independently to the probability.

Equal importance of features.

The conditional distribution of features is known or can be estimated.

** What are the advantages and disadvantages of Naïve Bayes**

Advantages:

Fast and simple.

Works well with high-dimensional data.

Requires relatively small training data.

Disadvantages:

Strong independence assumption is rarely true.

Can perform poorly if features are highly correlated.

**Why is Naïve Bayes a good choice for text classification**

Because text features (words) can be treated as conditionally independent, and the model handles large vocabularies efficiently with good performance despite the naive assumption.

 **Compare SVM and Naïve Bayes for classification tasks**

SVM often provides higher accuracy and better generalization for complex, high-dimensional, or non-linear data.

Naïve Bayes is faster, easier to implement, and works well for large-scale and text-based classification despite its simplicity.

** How does Laplace Smoothing help in Naïve Bayes?**

Laplace smoothing adds a small positive constant to feature counts to avoid zero probability for unseen features during prediction, ensuring more stable and robust probability estimates.

```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.datasets import load_iris, load_wine, load_breast_cancer, fetch_20newsgroups
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import (
    accuracy_score, mean_squared_error, mean_absolute_error, precision_recall_fscore_support,
    roc_auc_score, log_loss, confusion_matrix, precision_recall_curve, auc
)
from sklearn.svm import SVC, SVR
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier

import warnings
warnings.filterwarnings('ignore')

# #H: Train an SVM Classifier on the Iris dataset and evaluate accuracy
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
svm_iris = SVC()
svm_iris.fit(X_train, y_train)
print("Iris SVM Accuracy:", accuracy_score(y_test, svm_iris.predict(X_test)))

# ##: Train two SVM classifiers with Linear and RBF kernels on Wine dataset, compare accuracies
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, random_state=42)
svm_linear = SVC(kernel='linear')
svm_rbf = SVC(kernel='rbf')
svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)
print("Wine SVM Linear Kernel Accuracy:", accuracy_score(y_test, svm_linear.predict(X_test)))
print("Wine SVM RBF Kernel Accuracy:", accuracy_score(y_test, svm_rbf.predict(X_test)))

# #$: Train an SVR on a housing dataset and evaluate using MSE
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, random_state=42)
svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
print("SVR MSE:", mean_squared_error(y_test, y_pred))

# #%: Train an SVM Classifier with Polynomial Kernel and visualize decision boundary (2D example)
from sklearn.datasets import make_circles
X, y = make_circles(noise=0.2, factor=0.5, random_state=42)
svm_poly = SVC(kernel='poly', degree=3)
svm_poly.fit(X, y)
def plot_decision_boundary(clf, X, y):
    h = .02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
    plt.title("SVM Polynomial Kernel Decision Boundary")
    plt.show()
plot_decision_boundary(svm_poly, X, y)

# #!: Train Gaussian Naive Bayes on Breast Cancer dataset and evaluate accuracy
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
print("GaussianNB Breast Cancer Accuracy:", accuracy_score(y_test, gnb.predict(X_test)))

# #;: Train Multinomial Naive Bayes on 20 Newsgroups for text classification
newsgroups_train = fetch_20newsgroups(subset='train', categories=['alt.atheism', 'sci.space'])
newsgroups_test = fetch_20newsgroups(subset='test', categories=['alt.atheism', 'sci.space'])
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(newsgroups_train.data)
X_test = vectorizer.transform(newsgroups_test.data)
mnb = MultinomialNB()
mnb.fit(X_train, newsgroups_train.target)
print("MultinomialNB 20 Newsgroups Accuracy:", accuracy_score(newsgroups_test.target, mnb.predict(X_test)))

# ?= Train SVM with different C values and compare decision boundaries visually (2D example)
X, y = make_circles(noise=0.2, factor=0.5, random_state=42)
C_values = [0.1, 1, 10]
plt.figure(figsize=(15,5))
for i, C in enumerate(C_values):
    svm_c = SVC(kernel='rbf', C=C)
    svm_c.fit(X, y)
    plt.subplot(1, len(C_values), i+1)
    plot_decision_boundary(svm_c, X, y)
    plt.title(f'SVM Decision Boundary C={C}')

# %= Train Bernoulli Naive Bayes on binary feature dataset
from sklearn.datasets import make_multilabel_classification
X, y = make_multilabel_classification(n_samples=500, n_features=10, n_classes=1, random_state=42)
y = y.ravel()  # single label for BernoulliNB
bnb = BernoulliNB()
bnb.fit(X, y)
print("BernoulliNB Accuracy:", accuracy_score(y, bnb.predict(X)))

# = Apply feature scaling before SVM and compare with unscaled
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
svm_unscaled = SVC()
svm_unscaled.fit(X_train, y_train)
acc_unscaled = accuracy_score(y_test, svm_unscaled.predict(X_test))
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svm_scaled = SVC()
svm_scaled.fit(X_train_scaled, y_train)
acc_scaled = accuracy_score(y_test, svm_scaled.predict(X_test_scaled))
print(f"SVM accuracy unscaled: {acc_unscaled}, scaled: {acc_scaled}")

# *= Train Gaussian Naive Bayes and compare predictions before and after Laplace smoothing
gnb_no_smooth = GaussianNB()
gnb_no_smooth.fit(X_train, y_train)
pred_no_smooth = gnb_no_smooth.predict(X_test)

gnb_smooth = GaussianNB(var_smoothing=1e-9)
gnb_smooth.fit(X_train, y_train)
pred_smooth = gnb_smooth.predict(X_test)

print("GaussianNB accuracy before smoothing:", accuracy_score(y_test, pred_no_smooth))
print("GaussianNB accuracy after smoothing:", accuracy_score(y_test, pred_smooth))

# = Train SVM with GridSearchCV for C, gamma, kernel tuning
param_grid = {'C':[0.1,1,10],'gamma':[0.01,0.1,1],'kernel':['rbf','linear']}
grid = GridSearchCV(SVC(), param_grid, cv=3)
grid.fit(X_train_scaled, y_train)
print("Best GridSearchCV Params:", grid.best_params_)
print("Best GridSearchCV Accuracy:", grid.best_score_)

# = Train SVM on imbalanced dataset with class weighting
from sklearn.datasets import make_classification
X, y = make_classification(n_classes=2, weights=[0.9,0.1], n_samples=1000, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
svm_no_weight = SVC()
svm_no_weight.fit(X_train, y_train)
acc_no_weight = accuracy_score(y_test, svm_no_weight.predict(X_test))
svm_weighted = SVC(class_weight='balanced')
svm_weighted.fit(X_train, y_train)
acc_weighted = accuracy_score(y_test, svm_weighted.predict(X_test))
print(f"Accuracy without class weight: {acc_no_weight}")
print(f"Accuracy with class weight: {acc_weighted}")

# = Naive Bayes for spam detection (simple example with SMS Spam dataset from sklearn datasets)
# from sklearn.datasets import fetch_openml
# sms = fetch_openml(name='sms_spam', version=1, as_frame=True)
# X = sms.data['text']
# y = (sms.target == 'spam').astype(int)
# vectorizer = CountVectorizer()
# X_vect = vectorizer.fit_transform(X)
# X_train, X_test, y_train, y_test = train_test_split(X_vect, y, random_state=42)
# bnb = BernoulliNB()
# bnb.fit(X_train, y_train)
# print("Spam detection BernoulliNB Accuracy:", accuracy_score(y_test, bnb.predict(X_test)))

# = Train SVM and Naive Bayes on same dataset and compare accuracy
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
svm = SVC()
svm.fit(X_train, y_train)
nb = GaussianNB()
nb.fit(X_train, y_train)
print("SVM Accuracy:", accuracy_score(y_test, svm.predict(X_test)))
print("Naive Bayes Accuracy:", accuracy_score(y_test, nb.predict(X_test)))

# >= Feature selection before Naive Bayes and compare results
selector = SelectKBest(chi2, k=2)
X_new = selector.fit_transform(iris.data, iris.target)
X_train, X_test, y_train, y_test = train_test_split(X_new, iris.target, random_state=42)
nb = GaussianNB()
nb.fit(X_train, y_train)
print("Naive Bayes with feature selection Accuracy:", accuracy_score(y_test, nb.predict(X_test)))

# <= Train SVM with OvR and OvO on Wine dataset and compare accuracy
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, random_state=42)
ovr = OneVsRestClassifier(SVC())
ovo = OneVsOneClassifier(SVC())
ovr.fit(X_train, y_train)
ovo.fit(X_train, y_train)
print("OvR Accuracy:", accuracy_score(y_test, ovr.predict(X_test)))
print("OvO Accuracy:", accuracy_score(y_test, ovo.predict(X_test)))

# ?= Train SVM with Linear, Polynomial, and RBF kernels on Breast Cancer and compare accuracy
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)
kernels = ['linear', 'poly', 'rbf']
for k in kernels:
    svm = SVC(kernel=k)
    svm.fit(X_train, y_train)
    print(f"SVM kernel={k} Accuracy:", accuracy_score(y_test, svm.predict(X_test)))

# %= Stratified K-Fold Cross-Validation with SVM and average accuracy
skf = StratifiedKFold(n_splits=5)
accuracies = []
for train_index, test_index in skf.split(iris.data, iris.target):
    X_train, X_test = iris.data[train_index], iris.data[test_index]
    y_train, y_test = iris.target[train_index], iris.target[test_index]
    svm = SVC()
    svm.fit(X_train, y_train)
    accuracies.append(accuracy_score(y_test, svm.predict(X_test)))
print("Stratified K-Fold average accuracy:", np.mean(accuracies))

# = Naive Bayes with different prior probabilities and performance comparison
priors = [None, [0.6, 0.4, 0.0], [0.33, 0.33, 0.34]]
for p in priors:
    gnb = GaussianNB(priors=p)
    gnb.fit(iris.data, iris.target)
    print(f"GaussianNB with priors={p} Accuracy:", accuracy_score(iris.target, gnb.predict(iris.data)))

# *= Recursive Feature Elimination (RFE) before training SVM and compare accuracy
selector = RFE(SVC(kernel='linear'), n_features_to_select=2)
X_new = selector.fit_transform(iris.data, iris.target)
X_train, X_test, y_train, y_test = train_test_split(X_new, iris.target, random_state=42)
svm = SVC()
svm.fit(X_train, y_train)
print("SVM after RFE Accuracy:", accuracy_score(y_test, svm.predict(X_test)))

# = Train SVM and evaluate Precision, Recall, and F1-Score
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
svm = SVC()
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
print(f"Precision: {precision}, Recall: {recall}, F1-score: {f1}")

# = Train Naive Bayes and evaluate using Log Loss
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_prob = gnb.predict_proba(X_test)
print("GaussianNB Log Loss:", log_loss(y_test, y_prob))

# = Train SVM and visualize Confusion Matrix using seaborn
svm = SVC()
svm.fit(X_train, y_train)
cm = confusion_matrix(y_test, svm.predict(X_test))
sns.heatmap(cm, annot=True, fmt='d')
plt.title("SVM Confusion Matrix")
plt.show()

# = Train SVR and evaluate using Mean Absolute Error (MAE)
svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
print("SVR MAE:", mean_absolute_error(y_test, y_pred))

# >= Train Naive Bayes and evaluate using ROC-AUC score (binary classification)
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_score = gnb.predict_proba(X_test)[:, 1]
print("GaussianNB ROC-AUC score:", roc_auc_score(y_test, y_score))

# <= Train SVM and visualize Precision-Recall Curve
svm = SVC(probability=True)
svm.fit(X_train, y_train)
y_scores = svm.predict_proba(X_test)[:, 1]
precision, recall, _ = precision_recall_curve(y_test, y_scores)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (SVM)")
plt.show()
```
